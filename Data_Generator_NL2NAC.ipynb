```python
from tensorflow import keras
import random

for reports in range(100):
  conv_layer_counts = [4,5,6,6,6,7]
  conv_filters = [32,64,128,256,512,1024]
  conv_strides = [1,1,1,1,1,1,1,2]
  dense_layer_counts = [1,2,3,4]
  dense_units = [128,256,512,1024,2048]

  conv_layer_counts_str = ["4", "5", "6", "6", "6", "7"]
  conv_filters_str = ["32","64","128","256","512","1024"]
  conv_strides_str = ["stride-1", "stride-1", "stride-1", "stride-1", "stride-1", "stride-1", "stride-1", "stride-2"]
  dense_layer_counts_str = ["1","2","3","4"]
  dense_units_str = ["128","256","512","1024","2048"]

  architecture_string = "" # to be mapped to NL decriptions? title matching for now?

  code_string = ""
  

  # Demo Computer Vision architecture
  inputs = keras.Input(shape=(32,32,3))
  code_string += "inputs = keras.Input(shape=(32,32,3)) <eos>"
  x = keras.layers.Conv2D(64, 3, strides=1, padding='same', activation='relu')(inputs)
  code_string += "x = keras.layers.Conv2D(64, 3, strides=1, padding='same', activation='relu')(inputs) <eos>"


  c_l_idx = random.randint(0, len(conv_layer_counts)-1)
  architecture_string += conv_layer_counts_str[c_l_idx]
  architecture_string += "-"

  filter_size = 32 # initialization

  for ii in range(conv_layer_counts[c_l_idx]):
    c_f_idx = random.randint(0, len(conv_filters)-1)
    architecture_string += conv_filters_str[c_f_idx]
    architecture_string += "-"
    stride_idx = random.randint(0, len(conv_strides)-1)
    architecture_string += conv_strides_str[stride_idx]
    architecture_string += "-"

    stride = conv_strides[stride_idx]
    if stride == 1:
      filter_size -= 2
    else:
      filter_size /= 2
    
    if filter_size < 2:
      architecture_string += "-filter_exit-"
      break
      
    x = keras.layers.Conv2D(conv_filters[c_f_idx], 3, strides=conv_strides[stride_idx], padding="valid", activation="relu")(x)
    code_string += "x = keras.layers.Conv2D(" + str(conv_filters[c_f_idx]) + ", 3, strides=" + str(conv_strides[stride_idx]) + ", padding='valid', activation='relu')(x) <eos>"

  x = keras.layers.Flatten()(x)
  code_string += "x = keras.layers.Flatten()(x)"

  d_l_idx = random.randint(0, len(dense_layer_counts)-1)
  architecture_string += dense_layer_counts_str[d_l_idx]
  architecture_string += "-"

  for jj in range(dense_layer_counts[d_l_idx]):
    d_u_idx = random.randint(0, len(dense_units)-1)
    architecture_string += dense_units_str[d_u_idx]
    architecture_string += "-"
    x = keras.layers.Dense(dense_units[d_u_idx], activation="relu")(x)
    code_string += "x = keras.layers.Dense("+str(dense_units[d_u_idx]) + ", activation='relu')(x) <eos>"

  outputs = keras.layers.Dense(10, activation="softmax")(x)
  code_string += "outputs = keras.layers.Dense(10, activation='softmax')(x) <eos>"
  code_string += "model = keras.Model(inputs=inputs, outputs=outputs) <eos>"


  model = keras.Model(inputs=inputs, outputs = outputs)

  if reports % 100 == 0:
    print(reports)

  # save parameter counts -- needs to be parsed
  param_folder = "Parameter_Counts/"
  with open(param_folder + architecture_string + '.txt','w') as fh:
      # Pass the file handle in as a lambda function to make it callable
      model.summary(print_fn=lambda x: fh.write(x + '\n'))

  # save code to render architecture      
  code_folder = "Code_Strings/"
  with open(code_folder + architecture_string + '.txt', 'w') as fh:
    fh.write(code_string)

  del model
```
